```python

# !. Load .csv file
import pandas as pd
import numpy as np
import re
from IPython.display import display as ds
df = pd.read_csv(r"C:\Users\maste\OneDrive\Desktop\aus_ccx_21.csv")
print(ds(df))
```

```python
# see / explore the data

df.dtypes
```

```python
# 2. Removal of leading and trailing space from  all string col.

df[df.select_dtypes(include='object').columns] = df[df.select_dtypes(include='object').columns].astype('string').map(str.strip, na_action='ignore')

df
```

```python
# 3.Email - change to lower case , validate , replace invalid with NaN (as per email)

df['email'] = df['email'].str.lower()

valid_email = df['email'].str.match(r'^[\w\.-]+@[\w\.-]+\.\w+$' , na = False)

df.loc[~ valid_email , 'email' ] = np.nan

df
```

```python
# 4. Dates Correction and Replace to NaN ( as per email) 

df['join_date'] = df['join_date'].replace('Not Available' , np.nan)

df['join_date'] = pd.to_datetime(df['join_date'] , errors = 'coerce' , dayfirst = True)

df
```

```python

# 5 . Salary remove $ and USD change to empty string , change to numeric

df['salary'] = df['salary'].replace(r'[\$,USD\s]','',regex=True)

df['salary'] = pd.to_numeric(df['salary'] , errors='coerce')

print(ds(df))

print(df.dtypes)
```

```python
# 5 . Remarks Col -  remove trailing and leading space , also double and single space between the string
# as per email

df['remarks'] = df['remarks'].mask(df['remarks'].isna() , np.nan)

df['remarks'] = df['remarks'].str.strip()

df['remarks'] = df['remarks'].str.replace('  ' ,' ',regex = False)

df
```

```python

# 6. Sales = remove $ and leave <NA> as it is.

df['sales'] = df['sales'].replace(r'[\$,]' , '' , regex = True)

df['sales'] = pd.to_numeric(df['sales'] , errors = 'coerce')

df
```
